```
def dailyData(symbol, apiKey):
    symbolURL = "symbol=" + str(symbol)
    apiURL = "apikey=" + str(apiKey)
    
    completeURL = 'https://www.alphavantage.co/query?' +'function=TIME_SERIES_DAILY' + '&' + symbolURL + '&' + apiURL + '&datatype=csv' +'&outputsize=full'
    data = pd.read_csv(completeURL)
    #data.to_csv(filename = )
    return data
```

```
def getEMA(symbol, apiKey):
    symbolURL = "symbol=" + str(symbol)
    apiURL = "apikey=" + str(apiKey)
    completeURL = 'https://www.alphavantage.co/query?' + 'function=EMA'+ '&' + symbolURL + '&' + apiURL + '&datatype=csv' +'&interval=daily' \
    +'&time_period=14' +'&series_type=close'
    data = pd.read_csv(completeURL) 
    return data
```

```
def getRSI(symbol, apiKey):
    symbolURL = "symbol=" + str(symbol)
    apiURL = "apikey=" + str(apiKey)
    completeURL = 'https://www.alphavantage.co/query?' + 'function=RSI'+ '&' + symbolURL + '&' + apiURL + '&datatype=csv' +'&interval=daily' \
    +'&time_period=14' +'&series_type=close'
    data = pd.read_csv(completeURL) 
    return data
```

```
def getMFI(symbol, apiKey):
    symbolURL = "symbol=" + str(symbol)
    apiURL = "apikey=" + str(apiKey)
    completeURL = 'https://www.alphavantage.co/query?' + 'function=MFI'+ '&' + symbolURL + '&' + apiURL + '&datatype=csv' +'&interval=daily' \
    +'&time_period=14' +'&series_type=close'
    data = pd.read_csv(completeURL) 
    return data
```
