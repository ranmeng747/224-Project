```python
#mainURL that will be appended to return our requests

#API key that will be needed to authenticate


#For daily activity
'''
REQUEST NEEDS:
function: will ALWAYS be equal to 'TIME_SERIES_DAILY' for daily
symbol: user defined stock ticker
outputsize: default is 'compact' which only returns 100 data points, 
otherwise
we can define it as 'full' to get up to 20 years worth
datatype: default is 'json', but we can also request 'csv'
'''

#loads a json object of based on what you input in dailyData(SYMBOL)
def dailyData(symbol, apiKey):
    symbolURL = "symbol=" + str(symbol)
    apiURL = "apikey=" + str(apiKey)
    
    completeURL = 'https://www.alphavantage.co/query?' +'function=TIME_SERIES_DAILY' + '&' + symbolURL + '&' + apiURL + '&datatype=csv' +'&outputsize=full'
    data = pd.read_csv(completeURL)
    
    return data
```

```python
keys  = ['5KY0XZVM9KMKHMZX', 'J3CZ6JQALA5SJD18','30U8NSUYDXM4BC58']
```

```python
SMA = {}
EMA ={}
RSI = {}
MFI = {} 
MACD = {}
STOCH = {}
AD = {}
ADX = {}
OBV = {}
AROON = {}
CCI = {}


for stock in stocks:
    SMA[stock] = getSMA(stock, keys[0])  
    time.sleep(12)
    
    EMA[stock] = getEMA(stock, keys[0])
    time.sleep(12)
    
    RSI[stock] = getRSI(stock, keys[0])
    time.sleep(12)
    
    MFI[stock] = getMFI(stock, keys[0])
    time.sleep(12)
    
    MACD[stock] = getMACD(stock, keys[0])  
    time.sleep(12)
    
    STOCH[stock] = getSTOCH(stock, keys[0])
    time.sleep(12)
    
    AD[stock] = getAD(stock, keys[0])
    time.sleep(12)
    
    ADX[stock] = getADX(stock, keys[0])
    time.sleep(12)
    
    OBV[stock] = getOBV(stock, keys[0])
    time.sleep(12)
    
    AROON[stock] = getAROON(stock, keys[0])
    time.sleep(12)
    
    CCI[stock] = getCCI(stock, keys[0])
    time.sleep(12)

    print(stock, ' done') 

# it takes like 20 min
```

```python
alldata = {}
for stock in stocks:
    alldata[stock] = pd.concat([SMA[stock], EMA[stock], RSI[stock], MFI[stock], MACD[stock], STOCH[stock], 
                               AD[stock], ADX[stock], OBV[stock], AROON[stock], CCI[stock]], axis = 1).dropna() #dictionary value
    
    alldata[stock] = alldata[stock].loc[:,~alldata[stock].columns.duplicated()]
```
